```python
3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
   Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]
```

```python
series1 = pd.Series([2, 4, 6, 8, 10])
print("Series 1:",series1)
print()
series2 = pd.Series([1, 3, 5, 7, 9])
print("Series 2:",series2)
print()

add = series1+series2
print("Addition:",add)
print()


sub = series1-series2
print("Subctraction:",sub)
print()


mul = series1*series2
print("Multiplication:",mul)
print()

div = series1/series2
print("Division:",div)
print()
```

```python
9. i) Create a list of empids and names (10 employees).
   ii) Convert list into Series.
   iii) Print type of Series.
   iv) Convert Series into DataFrame.
   v) a) Display all records.
      b) Display first five records.
      c) Display last five records.
      d) Display first two and last two records.
      e) Display name of 5th employee.
      f) Display record of 4th employee.
   vi) Save the record in csv file (MyRecord.csv)
```

```python
#i) Create a list of empids and names (10 employees).
names = names = ["Olivia", "Liam", "Ava", "Noah", "Isabella", "Sophia", "Jackson", "Mia", "Aiden",
                 "Charlotte"]
print("Names:\n",names)
print("*"*90)

empids = [7186, 3036, 7937, 8375, 8994, 3620, 9127, 5635, 1269, 4226]
print("Employee Id's:\n",empids)

```

```python
#ii) Convert list into Series.
names_series = pd.Series(names)
print("Series 1:\n",names_series)
print()
empids_series = pd.Series(empids)
print("Series 2:\n",empids_series)
```

```python
#iii) Print type of Series.
print("Type of Series 1:",type(names_series))
print()
print("Type of Series 2:",type(empids_series))
```

```python
#iv) Convert Series into DataFrame.
df = pd.DataFrame([names_series,empids_series])
df = df.T
columns_name = {0:"Name",1:"Emp ID"}
df = df.rename(columns=columns_name)
df
```

```python
#v) a) Display all records.
print("DataFrame:\n",df)
print("*"*90)
```

```python
#   b) Display first five records.
print(df.head())
print("*"*90)
```

```python
#   c) Display last five records.
print(df.tail())
print("*"*90)
```

```python
#   d) Display first two and last two records.
print(df.loc[:1])
print("*"*90)
print(df.iloc[-2:])
```

```python
#    e) Display name of 5th employee.
print(df["Name"].iloc[5])
```

```python
#   f) Display record of 4th employee.
print(df.loc[4])
```

```python
#   vi) Save the record in csv file (MyRecord.csv)
df.to_csv("MyRecord.csv")
```
