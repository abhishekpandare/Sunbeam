```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Plot various graphs between empid and salary from myrecord.csv file.

```python
df = pd.read_csv("MyRecord.csv")
df.drop("Unnamed: 0",axis=1,inplace=True)
df.head()
```

```python
#data
name = df["Name"]
emp_id = df["Emp ID"]
```

```python
#Scatter Plot: Visualizes the relationship between two variables by plotting individual data points
#as dots on a graph.

#draw scatter plot
plt.scatter(name,emp_id,color = "red",label="Emp ID")
#customize the ply
plt.xlabel("Names")
plt.ylabel("Emp ID")
plt.title("Name and their id")
plt.legend()
plt.show()
```

```python
# Line Plot: Shows the relationship between two variables by connecting the data points with straight 
#     lines.
#draw line graph
plt.plot(name,emp_id,)
#customize the graph
plt.xlabel("Names")
plt.ylabel("Emp ID")
plt.title("Names and their ID's")
plt.show()
```

```python
#Bar Chart: Displays categorical data using rectangular bars. Useful for comparing values across 
#different categories.
#draw bar chart
plt.bar(name,emp_id,)
#customize the graph
plt.xlabel("Name")
plt.ylabel("Emp Id")
plt.title("Name and their Ids")
plt.show()
```

```python
#Histogram: Represents the distribution of a continuous variable by dividing it into bins and
#displaying the frequency of observations in each bin.
```

```python
#Pie Chart: Represents data as a circular chart divided into slices, where each slice represents 
#    a proportion of the whole.
plt.pie(emp_id)
plt.show()
```

```python
#i) Load Dataset by using following : df=sns.load_dataset('tips')
df=sns.load_dataset('tips')
df
```

```python
#ii) Plot stripplot between sex and tip with time as color.
sns.stripplot(x='sex',y='tip',data=df,hue='time')
```

```python
#iii) Plot swarmplot between day and tip with smoker as color.
sns.swarmplot(data= df,y='day',x='tip',hue='smoker')
```

```python
#iv) Plot swarmplot between total_bill and day with color as sex.
sns.swarmplot(data = df,x='total_bill',y='day',hue='sex')
```

```python
#v) Plot Boxplot of sex and total_bill.
sns.boxplot(data=df,x='sex',y='total_bill')
```

```python
#vi) Plot Violinplot between day and tip.
sns.violinplot(data=df,x="day",y="tip")
```

```python
#vii) Plot subplots which include violin and boxplot between sex and tip.
plt.subplot(1,3,1)
sns.violinplot(x='sex',y='tip',data=df)
plt.subplot(1,3,2)
sns.boxplot(x='sex',y='tip',data=df)
plt.subplot(1,3,3)
sns.stripplot(x='sex',y='tip',data=df)
```

```python
#viii) Plot subplots which include stripplot and swarmplot between sex and total_bill.
plt.subplot(1,2,1)
sns.stripplot(x='sex',y='total_bill',data=df)
plt.subplot(1,2,2)
sns.swarmplot(x='sex',y='total_bill',data=df)
```

```python
#ix) Plot subplots which include stripplot violinplot and boxplot between sex and total_bill.
plt.subplot(1,3,1)
sns.stripplot(x='sex',y='total_bill',data=df)
plt.subplot(1,3,2)
sns.violinplot(x='sex',y='total_bill',data=df)
plt.subplot(1,3,3)
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
#x) Plot boxplot over violinplot between day and tip.
plt.subplot(1,1,1)
sns.boxplot(x='day',y='tip',data=df)
plt.subplot(1,1,1)
sns.violinplot(x='day',y='tip',data=df)
```
